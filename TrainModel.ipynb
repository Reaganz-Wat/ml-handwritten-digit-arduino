# Building the Model

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
# create a classifier object
classifier = Sequential()

# First layer
classifier.add(Dense(units = 600, activation = "relu", kernel_initializer = 'uniform', input_dim = 784))

# Second layer
classifier.add(Dense(units = 400, activation = "relu", kernel_initializer = 'uniform'))

# Third layer
classifier.add(Dense(units = 200, activation = "relu", kernel_initializer = 'uniform'))

# Fourth layer
classifier.add(Dense(units = 10, activation = "sigmoid", kernel_initializer = 'uniform'))
```

### Compute the gradient descent algorithm to minimize the loss

```python
classifier.compile(optimizer='sgd', loss='mean_squared_error', metrics=['accuracy'])
```

### Here we start training by giving a batch size of 10, and epock of 10

```python
classifier.fit(x, y, batch_size=10, epochs=10)
```

# Make Predictions

```python
test_data = pd.read_csv('test.csv')
```

```python
y_predict = classifier.predict(test_data)
```
